[문제 2] 여러 숫자를 입력받아 그 합계를 반환하는 함수 sum_numbers를 작성하세요. 이 때, 가변 매개변수를 사용하세요.

```python
def sum_numbers():
  numbers = input("숫자들을 공백으로 입력하세여: ").split()
  total = 0
  for num in numbers:
    total += int(num)
  return total
print(sum_numbers())
```

[문제 6] 0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_numbers(n):
    for i in range(n + 1):
        if i % 2 == 1:
            yield i
odd = input("숫자를 입력하세요: ")
for num in odd_numbers(int(odd)):
    print(num)
```

[문제 8] 사용자로부터 3개의 숫자를 입력받아, 이를 튜플로 만든 후 최소값과 최대값을 출력하는 함수 min_max를 작성하세요.

```python
def min_max():
  numbers = input("3개의 숫자를 입력하세요: ").split()
  numbers = tuple(map(int, numbers))
  print(f"최소값: {min(numbers)}")
  print(f"최대값: {max(numbers)}")
min_max()
```

[문제 9] 주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

```python
def is_even(num):
  if num % 2 == 0:
    return True
  else:
    return False
num = int(input("숫자를 입력하세요: "))
print(is_even(num))
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n - 1)
num = int(input("숫자를 입력하세요: "))
print(factorial(num))
```

[문제 11] 문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

```python
def print_lengths(strings):
  for string in strings:
    print(f"{string}: {len(string)}")
strings = input("문자열을 입력하세요: ").split()
print_lengths(strings)
```

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(s1, s2):
  if len(s1) >= len(s2):
    return s1
  else:
    return s2
s1 = input("문자열을 입력하세요: ")
s2 = input("문자열을 입력하세요: ")
print(longer_string(s1, s2))
```
