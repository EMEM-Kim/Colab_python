__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다.

마법을 사용하려면 지능이 50 이상이어야 합니다.

마법사의 지능이 50 이상인지 확인하세요.

```python
wizard = {'strength': 10, 'dexterity': 20, 'intelligence': 80, 'health': 50}
warrior = {'strength': 80, 'dexterity': 30, 'intelligence': 20, 'health': 100}
thief = {'strength': 40, 'dexterity': 80, 'intelligence': 30, 'health': 60}
healer = {'strength': 20, 'dexterity': 40, 'intelligence': 70, 'health': 70}

if wizard['intelligence'] >= 50:
  print("마법을 사용할 수 있다!")
else:
  print("마법을 사용할 수 없다!")
```

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다.

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
if warrior['strength'] >=60 and warrior['health'] >= 80:
  print("전사는 강력한 공격을 할 수 있다!")
elif warrior['strength'] < 60:
  print("전사는 강력한 공격을 하기에는 힘이 약하다!")
else:
  print("전사는 강력한 공격을 하기에는 체력이 부족하다!")
```

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다.

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
if thief['dexterity'] > 70 or thief['strength'] > 50:
  print("도적은 몬스터를 회피할 수 있다!")
else:
  print("도적은 몬스터를 회피할 수 없다!")
```

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다.

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
if healer['intelligence'] >= 60 and healer['health'] >= 50:
  print("힐러는 파티의 체력을 회복시켰다!")
elif healer['intelligence'] < 60:
  print("힐러는 체력 회복 마법을 사용하기에는 지능이 부족하다!")
else:
  print("힐러는 체력 회복 마법을 사용하기에는 체력이 부족하다!")
```

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다.


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
wizard = {'strength': 10, 'dexterity': 20, 'intelligence': 80, 'health': 50, 'exp': 120}
warrior = {'strength': 80, 'dexterity': 30, 'intelligence': 20, 'health': 100, 'exp': 150}
thief = {'strength': 40, 'dexterity': 80, 'intelligence': 30, 'health': 60, 'exp': 110}
healer = {'strength': 20, 'dexterity': 40, 'intelligence': 70, 'health': 70, 'exp': 90}

party = {'wizard': wizard, 'warrior': warrior, 'thief': thief, 'healer': healer}

for name, party in party.items():
  if party['exp'] >=100:
    print(f"{name}은 레벨업이 가능하다!")
  else:
    lack_exp = 100 - party['exp']
    print(f"{name}은 레벨업을 하기에는 {lack_exp}만큼의 경험치가 부족하다")
```
