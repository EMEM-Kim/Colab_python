[문제 3]

사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done'이라고 입력할 때까지 계속 입력받으세요.

- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
numbers = []

while True:
    user_input = input("숫자를 입력하세요 (종료: done): ")

    if user_input.lower() == "done":
        break

    try:
        num = int(user_input)
        numbers.append(num)

    except ValueError:
        print(f"잘못된 값: {user_input}. 숫자를 입력해주세요.")

print("입력된 숫자 리스트:", numbers)
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요.

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time

def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"함수 실행 시간: {end_time - start_time:.2f}초")
        return result
    return wrapper

@timer_decorator
def example():
    time.sleep(1.5)
example()
````

````python
# mammals.py
%%writefile animals/mammals.py
class Dog:
    def __init__(self, name):
      self.name = name

    def info(self):
      return f"{self.name} is a dog."
````

````python
# birds.py
%%writefile animals/birds.py
class Eagle:
    def __init__(self, name):
      self.name = name

    def info(self):
      return f"{self.name} is an eagle."
````

````python
from animals.mammals import Dog
from animals.birds import Eagle

dog = Dog("영웅호걸")
eagle = Eagle("알테르")

print(dog.info())
print(eagle.info())
````
